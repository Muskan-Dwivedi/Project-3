From here we have graphs which are representing relation between different attributes such loyal and disloyal customer and on the basis of satisfaction

```python
loyal = df[df["satisfaction"]=="satisfied"]
disloyal =df[df["satisfaction"]=="neutral or dissatisfied"]
print("Loyal Customers:-",len(loyal))
print("disloyal Customers:-",len(disloyal))
```

```python
sns.countplot(x = "satisfaction",data = df)
plt.title("customer loyalty distribution")
plt.xlabel("Customer Status")
plt.ylabel("number of customer")
plt.show()
```

```python
sns.countplot(x = "Customer Type",hue = "satisfaction",data=df)
plt.title("Customer Type vs Satifaction")
plt.show()
```

```python
ratings = df.groupby("satisfaction")[[
   "Inflight entertainment",
   "Baggage handling",
   "Cleanliness"
]].mean()

ratings.plot(kind="bar")
plt.title("Average service Ratings")
plt.ylabel("Average Rating")
plt.xticks(rotation = 0)
plt.show()
```

```python
count = df["satisfaction"].value_counts()
plt.pie(count,labels=count.index,autopct = '%1.1f%%')
plt.title("Loyal vs Disloyal")
plt.show()
```
